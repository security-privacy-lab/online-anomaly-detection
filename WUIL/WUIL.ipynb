```python
import pandas as pd
```

```python
columns = ['ID', 'Date', 'Time', 'Session_ID', 'Depth', 'Path', 'Label']
dataset_benign = pd.read_csv("user1_log.txt", sep='|', header = None, names = columns )
dataset_benign.drop(['Date','Time'], axis = 1, inplace = True)
dataset_benign = dataset_benign.sort_values(by = 'Session_ID').reset_index(drop = True)
dataset_benign['Label'] = 0
dataset_benign
```

```python
malicious_logs1 = pd.read_csv('Attack1_log.txt', sep='|', header=None, names=columns)
malicious_logs2 = pd.read_csv('Attack2_log.txt', sep='|', header=None, names=columns)
malicious_logs3 = pd.read_csv('Attack3_log.txt', sep='|', header=None, names=columns)
combined_malicious_logs = pd.concat([malicious_logs1, malicious_logs2, malicious_logs3], ignore_index= True)
combined_malicious_logs
```

```python

combined_malicious_logs['Label'] = 1
combined_malicious_logs.drop(['Date','Time'], axis = 1, inplace = True)
combined_malicious_logs
```

```python

example_dataset = combined_malicious_logs.sample(frac = 0.4, random_state=123)

example_dataset = pd.concat([dataset_benign, combined_malicious_logs], ignore_index= True)
example_dataset.to_csv('Name.csv')
example_dataset
```

```python
edges = []
prev_path = None

for _, row in example_dataset.iterrows():
    current_path = row['Path']
    timestamp = row['Session_ID']
    label = row['Label']

    # Self-edge
    if prev_path == current_path:
        edges.append({
            'src_node' : current_path,
            'dst_node': current_path,
            'timestamp' : timestamp,
            'weight': 1, # Default weight
            'label' : label
        })
    elif prev_path is not None:
        edges.append({      
            'src_node' : prev_path,
            'dst_node': current_path,
            'timestamp' : row['Session_ID'],
            'weight': 1, # Default weight
            'label' : label # useful during evaluation, by default, it should be left as -
        })
    prev_path = current_path

edges_df = pd.DataFrame(edges)
edges_df
```

```python
edges = []
prev_path = None
edge_weights = {}  
for _, row in example_dataset.iterrows():
    current_path = row['Path']
    timestamp = row['Session_ID']
    label = row['Label']

    # Self-edge
    if prev_path == current_path:
        edge_key = (current_path, current_path)  
    elif prev_path is not None:
        edge_key = (prev_path, current_path)  
    else:
        prev_path = current_path
        continue  


    if edge_key in edge_weights:
        edge_weights[edge_key] += 1
    else:
        edge_weights[edge_key] = 1

    edges.append({
        'src_node': edge_key[0],
        'dst_node': edge_key[1],
        'timestamp': timestamp,
        'weight': edge_weights[edge_key],
        'label': label
    })

    prev_path = current_path  


edges_df = pd.DataFrame(edges)
edges_df
```

```python
edges_df[['src_node', 'dst_node', 'timestamp', 'weight', 'label']].to_csv(
    'sedanspot_inp2.csv', sep = ',', header = False, index = False
)
edges_df
```

Calculate the F1 Score and AUC

```python
from sklearn.metrics import f1_score, roc_auc_score, precision_recall_curve
import pandas as pd
threshold = 0.02
columns = ['src_node', 'dest_node', 'time', 'weight', 'label']
ground_truth = pd.read_csv("sedanspot_input.csv", header = None, names=columns)
anomaly_scores = pd.read_csv("output.csv", header=None, names=["score"])
results = pd.concat([ground_truth, anomaly_scores], axis=1)
results = results[["label", "score"]]
results["predicted"] = (results["score"] > threshold).astype(int)


f1 = f1_score(results["label"], results["predicted"])
print("F1 Score:", f1)

auc = roc_auc_score(results["label"], results["score"])
print("AUC Score:", auc)
```

```python
precision, recall, thresholds = precision_recall_curve(results['label'], results['score'])
f1_scores = 2 * (precision * recall) / (precision + recall)
best_threshold = thresholds[f1_scores.argmax()]
print("Best Threshold:", best_threshold)
print("Best F1 Score:", f1_scores.max())
```
